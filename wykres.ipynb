```python
import pandas as pd
import matplotlib.pyplot as plt

# Poprawiony link do surowego pliku CSV
url = "https://raw.githubusercontent.com/elli1zabeth/wyklad/main/Salary%20Data%20(2).csv"

# Prba wczytania z rnymi separatorami
try:
    data = pd.read_csv(url, sep=',')  # domylny przecinek
except:
    data = pd.read_csv(url, sep=';')  # jeli rednik

# Sprawdzenie kolumn
print(data.head())
print(data.columns)

# Zamy, e poprawne nazwy kolumn to 'Wiek' i 'Pensja' (moesz sprawdzi powyej)
# Jeli nazwy s inne, zmie je poniej
data = data.rename(columns={'Age': 'Wiek', 'Salary': 'Pensja'})

# Obliczamy redni pensj wedug wieku
srednia_pensja = data.groupby('Wiek')['Pensja'].mean()

# Tworzenie wykresu
plt.figure(figsize=(10, 5))
plt.plot(srednia_pensja.index, srednia_pensja.values, marker='o', linestyle='-')
plt.xlabel('Wiek')
plt.ylabel('rednia pensja')
plt.title('rednia pensja wedug wieku')
plt.grid()
plt.show()
```
