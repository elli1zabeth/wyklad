```python
import pandas as pd
import matplotlib.pyplot as plt

# Poprawiony link do surowego pliku CSV
url = "https://raw.githubusercontent.com/elli1zabeth/wyklad/main/Salary%20Data%20(2).csv"

# Próba wczytania z różnymi separatorami
try:
    data = pd.read_csv(url, sep=',')  # domyślny przecinek
except:
    data = pd.read_csv(url, sep=';')  # jeśli średnik

# Sprawdzenie kolumn
print(data.head())
print(data.columns)

# Załóżmy, że poprawne nazwy kolumn to 'Wiek' i 'Pensja' (możesz sprawdzić powyżej)
# Jeśli nazwy są inne, zmień je poniżej
data = data.rename(columns={'Age': 'Wiek', 'Salary': 'Pensja'})

# Obliczamy średnią pensję według wieku
srednia_pensja = data.groupby('Wiek')['Pensja'].mean()

# Tworzenie wykresu
plt.figure(figsize=(10, 5))
plt.plot(srednia_pensja.index, srednia_pensja.values, marker='o', linestyle='-')
plt.xlabel('Wiek')
plt.ylabel('Średnia pensja')
plt.title('Średnia pensja według wieku')
plt.grid()
plt.show()
```
